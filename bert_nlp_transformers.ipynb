```python
nlp=spacy.load('en_core_web_sm')
import spacy as spacy
import random as random
from collections import Counter
```

```python
def generate_mcqs(text,num_of_sentences):
  mcqs=[]
  doc=nlp(text)
  sentences=[sent.text for sent in doc.sents]
  stored_nouns=[token.text for token in nlp(text) if token.pos_=='NOUN']
  sel_sentences=random.sample(sentences,min(num_of_sentences,len(sentences)))
  stored_nouns=[token.text for token in nlp(text) if token.pos_=='NOUN']

  for sent in sel_sentences:
    words=nlp(sent)
    nouns=[token.text for token in words if token.pos_=='NOUN']
    if len(nouns)<2:
      continue
    counter=Counter(nouns)
    if counter:
      common_nouns=counter.most_common(1)[0][0]
      sentences=sent.replace(common_nouns,'_______')
      ans_choices=[common_nouns]
      sub=[common_nouns]
      while len(ans_choices)<4:
        if set(nouns) - set(ans_choices):
          distractor= random.choice(list(set(nouns) - set(ans_choices)))
        else:
          distractor= random.choice(list(set(stored_nouns) - set(ans_choices)))

        ans_choices.append(distractor)





      random.shuffle(ans_choices)
      ans=chr(64+ans_choices.index(common_nouns)+1)
      mcqs.append((
        sentences,ans_choices,ans
      ))

    # print(common_nouns)




  return mcqs









```

```python
from flask import Flask, render_template, request
from transformers import pipeline

app = Flask(__name__)

# Load NLP pipelines
sentiment_pipeline = pipeline('sentiment-analysis')
generation_pipeline = pipeline('text-generation')
translation_pipeline = pipeline('translation', model="Helsinki-NLP/opus-mt-fr-en")
summarization_pipeline = pipeline('summarization')
NER_pipeline = pipeline("ner", grouped_entities=True)


@app.route('/', methods=['GET', 'POST'])
def index():
    sent = None
    gen = None
    trans = None
    ner = None
    summ = None
    if request.method == 'POST':
        user_input = request.form.get('user_input')
        task = request.form.get('task')

        if task == 'sentiment':
            sent = sentiment_pipeline(user_input)
            return render_template('1_pipeline.html', sent=sent)
        elif task == 'generation':
            gen = generation_pipeline(user_input)
            return render_template('1_pipeline.html', gen=gen)
        elif task == 'translation':
            trans = translation_pipeline(user_input)
            return render_template('1_pipeline.html', trans=trans)
        elif task == 'summarization':
            summ = summarization_pipeline(user_input)
            return render_template('1_pipeline.html', summ=summ)
        elif task == 'named_entity_recognition':
            ner = NER_pipeline(user_input)
            return render_template('1_pipeline.html', ner=ner)



    return render_template('1_pipeline.html')

if __name__ == '__main__':
    app.run(debug=True)
```
